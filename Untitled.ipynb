```python
import pandas as pd
import matplotlib.pyplot as mp
from sklearn.model_selection import train_test_split as TTS,GridSearchCV as GSCV
from sklearn.metrics import classification_report as CR,accuracy_score as AS
from sklearn import set_config
import seaborn as sn
import matplotlib.pyplot as mp
import numpy as np
from sklearn.compose import ColumnTransformer as CT
from sklearn.pipeline import Pipeline as PP
from sklearn.preprocessing import StandardScaler as SS
set_config(display="diagram")
```

```python
from sklearn.ensemble import RandomForestClassifier as RFC
from xgboost import XGBClassifier as XGBC
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.naive_bayes import GaussianNB as GNB
```

```python
t.drop(["citric acid","free sulfur dioxide"],axis=1,inplace=True)
```

```python
t.head()
```

```python
t.quality.value_counts().plot.barh();
mp.xlabel("quality",fontsize=15);
```

```python
t.quality=np.where(t.quality>=7,1,0)
```

```python
x=t.drop("quality",axis=1)
x
```

```python
y=t.loc[:,"quality"]
y
```

```python
xta,xte,yta,yte=TTS(x,y,test_size=0.2,random_state=5)
```

```python
"RFC ="+str(AS(yte,RFC().fit(xta,yta).predict(xte)))
```

```python
pp=PP(
[
    ("ss",SS()),
    ("knn",KNN()),
])
"KNN ="+str(AS(yte,pp.fit(xta,yta).predict(xte)))
```

```python
pp1=PP(
[
    ("ss",SS()),
    ("svc",SVC()),
])
"SVC ="+str(AS(yte,pp1.fit(xta,yta).predict(xte)))
```

```python
"GNB ="+str(AS(yte,GNB().fit(xta,yta).predict(xte)))
```

```python
"XGBC ="+str(AS(yte,XGBC().fit(xta,yta).predict(xte)))
```

```python
p={
    "criterion":["gini","entropy"],
    "max_depth":np.arange(20),
    "min_samples_split":[2,3,4,5],
    "min_samples_leaf":[1,2,3,4],
    "bootstrap":[True,False]
    
}
```

```python
gscv=GSCV(RFC(),param_grid=p,scoring="accuracy",n_jobs=-1,cv=5)
```

```python
ppp=PP(
[
    ("SS",SS()),
    ("gscv",gscv)
]
,verbose=True)
```

```python
ppp.fit(xta,yta)
```

```python
AS(yte,ppp.predict(xte))
```

```python
ppp.named_steps["gscv"].best_params_
```
